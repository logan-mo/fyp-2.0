## Populate

```python
"""
Give the schema: 
{
    "detection_id": int,
    "camera_id": "str",
    "car_type": "str,
    "car_color": "str",
    "car_license_plate_number": "str",
    "date_time": "str",
}

where camera_id in ["CAM1", "CAM2", "CAM3"]
car_type in [
    "bus", "car_honda", "car_hyundai", "car_other", "car_suzuki", "car_toyota", "van_honda", "van_hyundai", "van_other", "van_suzuki", "van_toyota",
]
and color in: ["black", "blue", "green", "grey", "red", "silver", "white"]
Write a code that will populate the database with 1000 random records.


The random records must follow a few generic rules:

- The car type must be randomly selected from the list of car types.
- The car color must be randomly selected from the list of colors.
- The license plate number must be randomly generated. It must be in the format "ABC-1234" where A, B, and C are random uppercase letters and 1, 2, 3, and 4 are random digits.

Any car may be detected at CAM1, but some cars might also be detected at CAM2 (with a later date_time), and some cars might be detected at CAM1, CAM2 and CAM3 (with even later date_time).
"""

import random
import string
import datetime

def random_car_type():
    car_types = [
        "bus", "car_honda", "car_hyundai", "car_other", "car_suzuki", "car_toyota", "van_honda", "van_hyundai", "van_other", "van_suzuki", "van_toyota",
    ]
    return random.choice(car_types)

def random_car_color():
    colors = ["black", "blue", "green", "grey", "red", "silver", "white"]
    return random.choice(colors)

def random_license_plate_number():
    letters = string.ascii_uppercase
    digits = string.digits
    return f"{random.choice(letters)}{random.choice(letters)}{random.choice(letters)}-{random.choice(digits)}{random.choice(digits)}{random.choice(digits)}{random.choice(digits)}"

def random_date_time(greater_than = None):
    start = datetime.datetime(2021, 4, 23, 7, 0, 0)
    end = datetime.datetime(2021, 4, 23, 10, 0, 0)
    date_time = start + (end - start) * random.random()
    
    print(date_time, greater_than)
    
    if greater_than is None:
        return date_time
    
    while date_time < greater_than:
        date_time = start + (end - start) * random.random()
    return date_time

import mysql.connector

host = 'localhost'
user = 'root'
password = 'h1R1m!55'
database = "car_detection_db"

conn = mysql.connector.connect(host=host, user=user, password=password, database=database)

print(conn)

cursor = conn.cursor()

detection_id = 1
car_id = 1
for i in range(150):
    car_type = random_car_type()
    car_color = random_car_color()
    car_license_plate_number = random_license_plate_number()
    date_time1 = random_date_time()
    print(car_type, car_color, car_license_plate_number, date_time1)

    cursor.execute(f"INSERT INTO car_detection_db.detections (camera_id, car_type, car_color, car_licence_plate_number, date_time, detection_id) VALUES ('CAM1', '{car_type}', '{car_color}', '{car_license_plate_number}', '{date_time1}', {detection_id})")
    detection_id += 1
    if random.random() < 0.9:
        date_time2 = random_date_time(date_time1)
        
        cursor.execute(f"INSERT INTO car_detection_db.detections (camera_id, car_type, car_color, car_licence_plate_number, date_time, detection_id) VALUES ('CAM2', '{car_type}', '{car_color}', '{car_license_plate_number}', '{date_time2}', {detection_id})")
        detection_id += 1
        if random.random() < 0.8:
            date_time3 = random_date_time(date_time2)
                
            cursor.execute(f"INSERT INTO car_detection_db.detections (camera_id, car_type, car_color, car_licence_plate_number, date_time, detection_id) VALUES ('CAM3', '{car_type}', '{car_color}', '{car_license_plate_number}', '{date_time3}', {detection_id})")
            detection_id += 1
            
            
    cursor.execute(f"INSERT INTO car_detection_db.car_data (car_id, car_type, car_color, car_license_plate_number, car_licence_plate_color) VALUES ({car_id}, '{car_type}', '{car_color}', '{car_license_plate_number}', 'White')")
    car_id += 1
    conn.commit()
```
